```python
# used to parse values into the url
#Used to make requests
import urllib.request

# used to parse values into the url
import urllib.parse
#We want to acces mentioned below website
# http://pythonprogramming.net/?s=basics&submit=Search

url = 'https://pythonprogramming.net'
#we write(?s=basics&submit=Search) in a dictionary form
values = {'s':'basics','submit':'search'} #[Values is like a dictionary here]

data = urllib.parse.urlencode(values)
data = data.encode('utf-8') # data should be bytes
req = urllib.request.Request(url, data)
respond = urllib.request.urlopen(req)
respData = respond.read()

print(respData)
```

```python
# used to parse values into the url
#Used to make requests
import urllib.request

# used to parse values into the url
import urllib.parse

#import regular expression(re)
#we use re to extract paragraphs from the website source code
import re
#We want to acces mentioned below website
# http://pythonprogramming.net/parse-website-using-regular-expressions-urllib/'

url = 'http://pythonprogramming.net/parse-website-using-regular-expressions-urllib/'

req = urllib.request.Request(url)
resp = urllib.request.urlopen(req)
respData = resp.read()
```

```python
paragraphs = re.findall(r'<p>(.*?)</p>',str(respData))
for eachP in paragraphs:
    print(eachP)
```
